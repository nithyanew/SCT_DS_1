```python
import matplotlib.pyplot as plt
genders = ['Male', 'Female', 'Other']
counts = [50, 45, 5]
plt.bar(genders, counts)
plt.xlabel('Gender')
plt.ylabel('Number of People')
plt.title('Gender Distribution')
plt.show()
```
